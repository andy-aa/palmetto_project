```python
from sqlalchemy import create_engine
import pymysql
import pandas as pd
```

```python
SQL = '''
SELECT
    protocol_id.protocol_id as id,
    stb_mean_color.stb_mean_color as color,
    stb_mean_flex.stb_mean_flex as flex,
    stb_mean_length.stb_mean_length as length,
    stb_mean_tensity.stb_mean_tensity as tensity,
    stb_N.stb_N as N
FROM
    (
    SELECT DISTINCT
        protocol_id
    FROM
        estimation_dtl
) AS protocol_id
LEFT JOIN(
    SELECT
VALUE AS
    stb_mean_color,
    `protocol_id`
FROM
    estimation_dtl
WHERE
    code_name = 'stb_mean_color'
) AS stb_mean_color
ON
    protocol_id.protocol_id = stb_mean_color.protocol_id
LEFT JOIN(
    SELECT
VALUE AS
    stb_mean_flex,
    protocol_id
FROM
    estimation_dtl
WHERE
    code_name = 'stb_mean_flex'
) AS stb_mean_flex
ON
    protocol_id.protocol_id = stb_mean_flex.protocol_id
LEFT JOIN(
    SELECT
VALUE AS
    stb_mean_length,
    protocol_id
FROM
    estimation_dtl
WHERE
    code_name = 'stb_mean_length'
) AS stb_mean_length
ON
    protocol_id.protocol_id = stb_mean_length.protocol_id
LEFT JOIN(
    SELECT
VALUE AS
    stb_mean_tensity,
    protocol_id
FROM
    estimation_dtl
WHERE
    code_name = 'stb_mean_tensity'
) AS stb_mean_tensity
ON
    protocol_id.protocol_id = stb_mean_tensity.protocol_id
LEFT JOIN(
    SELECT
VALUE AS
    stb_N,
    protocol_id
FROM
    estimation_dtl
WHERE
    code_name = 'stb_N'
) AS stb_N
ON
    protocol_id.protocol_id = stb_N.protocol_id;
'''
```

```python
%%time
df = pd.read_sql(SQL, index_col='id', con=create_engine('mysql+pymysql://root:my_secret_password@db/TexlabDB'))
df.shape
```
